O stub de código fornecido lê um número inteiro, n, de STDIN. Para todos os inteiros não negativos, imprima.


```python
# Input
lista = []
n = int(input("Digite um número: "))

# Adiciona os números menores que n à lista
for i in range(0, n):
    lista.append(i)

# Calcula e imprie os quadrados dos números da lista
for j in lista:
  print(j*j)
```
