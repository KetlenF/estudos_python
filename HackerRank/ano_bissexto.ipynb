Um dia extra é adicionado ao calendário quase a cada quatro anos, a partir de 29 de fevereiro, e o dia é chamado de dia bissexto. Ele corrige o calendário para o fato de que nosso planeta leva aproximadamente 365,25 dias para orbitar o sol. Um ano bissexto contém um dia bissexto.

No calendário gregoriano, três condições são usadas para identificar anos bissextos:

O ano pode ser dividido igualmente por 4, é um ano bissexto, a menos que:
O ano pode ser dividido igualmente por 100, NÃO é um ano bissexto, a menos que:
O ano também pode ser dividido por 400. Então, é um ano bissexto.
Isso significa que, no calendário gregoriano, os anos 2.000 e 2.400 são anos bissextos, enquanto 1800, 1900, 2100, 2.200, 2300 e 2.500 NÃO são anos bissextos. Fonte

Tarefa

Dado um ano, determine se é um ano bissexto. Se for um ano bissexto, retorna o Boolean True, caso contrário, retorna False.

Observe que o stub de código fornecido lê de STDIN e passa argumentos para a função is_leap. É necessário apenas completar a função is_leap.

```python
# Funcao que verifica se o ano é bissexto
def is_leap(year):
  leap = False

  # Se ano é divisível por 400
  if (year%400 == 0):
    leap = True
  # Se ano é divisível por 100
  elif (year%100 == 0):
    leap = False
  # Se ano é divisível por 4
  elif (year%4 == 0):
    leap = True
  
  return leap

# Input year
year = int(input("Digite o ano para validação: "))
print(is_leap(year))
```
