O stub de código fornecido lê dois inteiros de STDIN, a e b. Adicione o código para imprimir três linhas onde:

A primeira linha contém a soma dos dois números.
A segunda linha contém a diferença dos dois números (primeiro - segundo).
A terceira linha contém o produto dos dois números.

```python
a = int(input("Primeiro Número: "))
b = int(input("Segundo Número: "))
    
# Soma dos dois numeros
soma = a + b
print(soma)

# Diferenca dos dois numeros
diferenca = a - b
print(diferenca)

# Produto dos dois numeros
produto = a * b
print(produto)
```
