O stub de código incluído lerá um número inteiro, n, de STDIN.

Sem usar nenhum método de string, tente imprimir o seguinte:
123 ... n

Observe que "..." representa os valores consecutivos intermediários.

```python
n = int(input("Digite ultimo numero da sequencia: "))
    
for i in range(1, n+1):
    print(i, end="")
```
